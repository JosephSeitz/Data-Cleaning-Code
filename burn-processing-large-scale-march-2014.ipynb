```python
def file_finder(path,f_type = ".dat" ):
    txt_files=[]
    all_files = os.listdir(path) #tells what directory the files are in
    for i in all_files: #reading only the .txt files 
        if i[-4:] == f_type:
            txt_files.append(i)
    txt_files.sort() #alphabetizes them
    return txt_files
```

```python
def apply_correction(df,u_fctr,v_fctr,w_fctr,m_speed,min_T,fill_nan,tc_corr = True):
    out_column_tc =['TC(0.5m)', 'TC(2m)', 'TC(3.5m)', 'TC(5m)', 'TC(6.5m)',\
                'TC(8m)', 'TC(9.5m)', 'TC(11m)']
    #max_T = 200
    ##For loop for all the sonics
    adjust_U = ['U']
    adjust_V = ['V']
    adjust_W = ['W']
    adjust_wind = adjust_U+adjust_V +adjust_W
    
    #adjust_temp = ['T_19m', 'T_9.5m', 'T_3m']
    for i in range(len(adjust_U)):
        df[adjust_U[i]] *= u_fctr
        df[adjust_V[i]] *= v_fctr
        df[adjust_W[i]] *= w_fctr
        
    #adjust_son = [['U(21.6m)','V(21.6m)','W(21.6m)','T(21.6m)'],['U(9.1m)','V(9.1m)','W(9.1m)','T(9.1m)'],['U(3.4m)','V(3.4m)','W(3.4m)','T(3.4m)']]

    indx = []
    for i in range(len(df)):
        if float(df['T'][i])< min_T or float(np.abs(df["U"][i])) > m_speed or \
                float(np.abs(df["V"][i]))> m_speed or float(np.abs(df["W"][i]))> m_speed :
            print("--"*15,"LINE:",i,"--"*15)
            print(df.iloc[[i]])
            indx.append(i)
            continue
        if tc_corr == True:
            
            for j in range(len(out_column_tc)):
                if df[out_column_tc[j]][i] < min_T:
                    print("--"*15,"LINE:",i,"--"*15)
                    print(df.iloc[[i]])
                    indx.append(i)
                    break    
    
    if len(indx) ==0:
        print("Data fits these limits")
    if len(indx)!= 0:
        #remove = input("Would you like to remove these lines? (y/n):")
        remove = "y"
        if remove.lower() == "y":
            df = df.drop(df.index[indx])      
   
    df.fillna(value=fill_nan, inplace=True)
    
    return df
```
